### Bar Graphs

```python
google=data.DataReader(name="GOOG", data_source="iex",start="2013-01-1",end="2017-01-01")
google.head(3)
```

```python
google["close"].plot(kind="bar")
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
google["close"].apply(rank_performance).value_counts().plot(kind="bar")
```

```python
google["close"].apply(rank_performance).value_counts().plot(kind="barh")
```

### Pie Charts

```python
apple=data.DataReader(name="AAPL",data_source='iex',start="2014-01-01",end="2016-01-01")
```

```python
apple.head(3)
```

```python
apple["close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 99.19456111111116:
        return "above average"
    else:
        return "below average"
```

```python
plt.style.use('ggplot')
apple["close"].apply(rank_performance).value_counts().plot(kind="pie",legend=True)
```

### Histograms

```python
def custom_round(stock_price):
    return int(stock_price)
```

```python
custom_round(535.97/100.0)*100
```

```python
google["high"].apply(custom_round).nunique()
```

```python
google["high"].apply(custom_round).value_counts().sort_index().plot(kind="hist",bins= 9)
```
